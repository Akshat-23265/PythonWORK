```python
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
items=[("A",90),("B",95),("C",95),]
k=2
result=top_k(items,k)
print(result)
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        if cat not in d:
            d[cat]={}
        if word not in d[cat]:
            d[cat][word]=0
        d[cat][word]+=1
    return d
pairs=[("sports","win"), ("sports","win"),("tech","ai")]
print(count_by_category(pairs))
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) / 2
    return int(total - sum(nums))
nums=[3,0,1]
print(missing_number(nums))
```

```python
def add_unique(x, seen=set()):
    if seen is None:
        seen=set()
    if x not in seen:
        seen.add(x)
    return list(seen)
add_unique(1)
add_unique(2)
```

```python
def moveing_average(stream,w):
    average=[]
    current_sum=0
    for i ,element in enumerate(stream):
        current_sum+=element
        if i>=w:
            current_sum-=stream[i-w]
        count=min(i+1,w)
        average.append(current_sum/count)
    return average
stream=[1,2,3,4,5]
window_size=3
print(moveing_average(stream,window_size))
```
